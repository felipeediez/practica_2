Importar librerias y almacenar archivo .csv en un Data Frame

*   librerias Matplotlib y Pandas
*   con pd.read_csv



```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv ('../data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')  # Leo archivo .csv y almaceno en variable df
```

Filtrar Data Frame:

*   Primero por la fecha mínima
*   Segundo por la fecha máxima



```python
df_1 = df [(df['Date'] >= "2005-01-01")]  # Primero filtro el Data Frame por todas las fechas mayores o iguales al 1 de Enero de 2005
df_period_selected = df_1 [(df['Date'] <= "2014-12-31")]  # Luego tomo el Data Frame anterior y le agrego el filtro de todas las fechas menores al 31 de Diciembre de 2014
df_period_selected.head()  # Data Frame que almacena las filas del periodo 2005 - 2014
```

Se comienza a trabajar con el Data Frame dentro del período indicado en la propuesta

```python
df_tmax = df_period_selected [df_period_selected ['Element'] == 'TMAX']  # Data Frame ya filtrado por fechas que solo contiene temperaturas máximas.
df_tmin = df_period_selected [df_period_selected ['Element'] == 'TMIN']  # Data Frame ya filtrado por fechas que solo contiene temperaturas minimas.

df_tmax = df_tmax.groupby(df_tmax ['Date'].dt.strftime( '%m-%d'))['Data_Value'].max()  # Se agrupan por dia del año todas las temperaturas máximas y se selecciona la máxima registrada en cada dia del año en el periodo indicado
df_tmin = df_tmin.groupby(df_tmin ['Date'].dt.strftime ('%m-%d'))['Data_Value'].min()  # Se agrupan por dia del año todas las temperaturas minimas y se selecciona la minima registrada en cada dia del año en el periodo indicado

df_tmax = df_tmax [df_tmax.index != '02-29']  # Nuevo Data Frame que incluye todas las fechas que no sean el 02-29
df_tmin = df_tmin[df_tmin.index != '02-29']  # Nuevo Data Frame que incluye todas las fechas que no sean el 02-29

tmax_celsius = df_tmax.values / 10  # Convertir de décimas de grado a grados Celsius
tmin_celsius = df_tmin.values / 10  # Convertir de décimas de grado a grados Celsius
```

Eje x


```python
dates_range = pd.date_range(start='2000-01-01', periods= 365, freq='D')  # Eje x que va de dia en dia, rango de fechas: todos los dias de un año
```

Ajustes eje x

```python
months = pd.date_range (start='2000-01-01', periods=12, freq='MS')  # Esta variable se utiliza para crear un rango de fechas mensuales para el eje x del gráfico
month_labels = [month.strftime('%b') for month in months]  # Bucle for para iterar sobre cada fecha en la lista 'months' y obtener el nombre abreviado del mes correspondiente
```

Data Frame Min y Max del 2015

```python
df_2015 = df[(df['Date'] >= '2015-01-01') & (df['Date'] <= '2015-12-31')]  # Filtrar los datos de 2015
df_tmax_2015 = df_2015[df_2015['Element'] == 'TMAX']  # Filtrar las temperaturas máximas de 2015
df_tmin_2015 = df_2015[df_2015['Element'] == 'TMIN']  # Filtrar las temperaturas mínimas de 2015
```

Dias en los que fue superado la temperatura máxima y minima del periodo 2005 - 2014

```python
records_superados_max = df_tmax_2015.groupby(df_tmax_2015['Date'].dt.strftime('%m-%d'))['Data_Value'].max() > df_tmax  # Se agrupan los dias con temperaturas maximas en 2015 y se establece si superan a las temperaturas del periodo 2005-2014. De ser asi el resultado será verdadero (True)
records_superados_min = df_tmin_2015.groupby(df_tmin_2015['Date'].dt.strftime('%m-%d'))['Data_Value'].min() < df_tmin  # Se agrupan los dias con temperaturas minimas en 2015 y se establece si superan a las temperaturas del periodo 2005-2014. De ser asi el resultado será verdadero (True)
```

Creación del gráfico

```python
  # Creación del grafico
plt.figure (figsize=(12, 6))
plt.plot (dates_range, tmax_celsius, color='red', linewidth=1, label='Temperatura Máxima Récord por Día (2005-2014)')
plt.plot (dates_range, tmin_celsius, color='blue', linewidth=1, label='Temperatura Mínima Récord por Día (2005-2014)')
plt.fill_between (dates_range, tmax_celsius, tmin_celsius, facecolor='lightgray')


plt.scatter(dates_range[records_superados_max], df_tmax[records_superados_max].values / 10, c='black', label='Récord Máximo de 2005-2014 fue superado en 2015')  # La función scatter se utiliza para mostrar en el gráfico los días específicos en los que se superaron los récords de temperaturas máximas en 2015 en comparación con el período 2005-2014
plt.scatter(dates_range[records_superados_min], df_tmin[records_superados_min].values / 10, c='green', label='Récord Mínimo de 2005-2014 fue superado en 2015')  # La función scatter se utiliza para mostrar en el gráfico los días específicos en los que se superaron los récords de temperaturas minimos en 2015 en comparación con el período 2005-2014

  # Detalles como el título, leyendas y etiquetas
plt.title ('Temperaturas Máximas y Mínimas Récord (2005-2014)')
plt.xlabel ('Fecha')
plt.ylabel ('Temperatura (°C)')
plt.legend ()
plt.xticks (months, month_labels, rotation=45)  # Se personaliza el eje x con etiquetas de mes, con cierta inclinación que ayudan a la visualización

  # Se muestra la gráfica
plt.show()
```
